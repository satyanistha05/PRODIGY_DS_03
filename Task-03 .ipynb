```python
column=df[["age", "campaign", "duration"]]
q1=np.percentile(column, 25)
q3=np.percentile(column, 75)
iqr=q3-q1
lower_bound=q1-1.5*iqr
upper_bound=q3+1.5*iqr
df[["age", "campaign", "duration"]]=column[(column>lower_bound) & (column<upper_bound)]
df.plot(kind="box", subplots=True, layout=(2,5), figsize=(20,10), color="#FF1493")
```

```python
high_corr_cols=["emp.var.rate", "euribor3m","nr.employed"]
```

```python
df1=df.copy()
df1.columns
```

```python
df1.drop(high_corr_cols, inplace=True, axis=1) 
df1.columns
```

```python
df1.shape
```

```python
from sklearn.preprocessing import LabelEncoder
ib= LabelEncoder()
df_encoded=df1.apply(ib.fit_transform)
df_encoded
```

```python
df_encoded["y"].value_counts()
```

```python
x=df_encoded.drop("y", axis=1)
y=df_encoded ["y"]
print(x.shape)
print(y.shape)
print(type(x))
print(type(y))
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=1)
print(x_train.shape) 
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
def eval_model(y_test, y_pred):
    acc=accuracy_score(y_test, y_pred)
    print("Accuracy score", acc)
    cm=confusion_matrix(y_test, y_pred)
    print("Confusion Matrix ",cm)
    crp=classification_report(y_test, y_pred)
    print("Clssification report ",crp)
```

```python
def mycore(model):
    train_score=model.score(x_train, y_train)
    test_score=model.score(x_test, y_test)
    print("Training score ", train_score)
    print("Testing score", test_score)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier(criterion="gini",max_depth=5, min_samples_split=10) 
dt.fit(x_train, y_train)
```

```python
ypred_dt1=dt.predict(x_test)
eval_model(y_test, ypred_dt1)
```

```python
from sklearn.tree import plot_tree
cn=["no", "yes"]
fn=x_train.columns
print(fn)
print(cn)
```

```python
plot_tree(dt,class_names=cn, filled=True) 
plt.show()
```

```python
dt1=DecisionTreeClassifier(criterion="entropy",max_depth=4,min_samples_split=15) 
dt1.fit(x_train, y_train)
```

```python
mycore(dt1)
```

```python
ypred_dt1=dt1.predict(x_test) 
eval_model(y_test, ypred_dt1)
```
